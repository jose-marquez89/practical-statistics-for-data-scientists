# Data and Sampling Distributions

## Long Tailed Distributions
- Data is not generally normally distributed
- extreme events are more likely to occur than would be predicted by the normal distribution

### Key Terms
- Skew: Where one tail of a distribution is longer than the other



```python
import os
import requests

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Pull some stock or crypto data to look at a QQ plot

POLYGON_KEY = os.environ["POLYGON_KEY"]
url = "https://api.polygon.io/v2/aggs/ticker/X:BTCUSD/range/1/day/2017-01-01/2021-01-01?adjusted=true&sort=asc&limit=120&apiKey="
req = url + POLYGON_KEY

res = requests.get(req)
prices = res.json()['results']
bitcoin = pd.DataFrame(prices)

bc = np.diff(np.log(bitcoin["c"]))
fig, ax = plt.subplots(figsize=(4,4))
stats.probplot(bc, plot=ax)

plt.show()
```
